```python
from transformers import pipeline
```

```python
imag_cap = pipeline("image-to-text", model = 'Salesforce/blip-image-captioning-large', device = 0)
```

```python
image = "/content/WhatsApp Image 2024-10-18 at 14.09.21_10f445af.jpg"
cap = imag_cap(image)
print(cap[0].get('generated_text'))
```

```python
vqa = pipeline("visual-question-answering", model = 'dandelin/vilt-b32-finetuned-vqa', device = 0)
```

```python
ques = input("Enter the question: ")
ans = vqa(image, ques)
ans
```
